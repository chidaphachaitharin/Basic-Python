```python
#Create a figure object called fig
#use add_axes to add new axis ax
#plot (x,y) on that axes and set the Labels and title to match the plot below
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
#plt.show()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('My title')
```
