```python
import pandas as pd
names = ['Bob','Jessica','Mary','John','Mel']
grades = [76,95,77,78,99]

GradeList = zip(names,grades)
columns = ['Names', 'Grades']
df = pd.DataFrame(data=GradeList,columns=columns)
df
```

```python
print('number od values:',df['Grades'].count())
print('Average:',df['Grades'].mean())
print('Standard deviation:',df['Grades'].std())
print('The max:',df['Grades'].max())
print('The min:',df['Grades'].min())
print('first quantile:',df['Grades'].quantile(.25))
print('second quantile:',df['Grades'].quantile(.50))
print('third quantile:',df['Grades'].quantile(.75))
```

```python
df['Grades'].median()
```

```python
df['Grades'].mode()
```

# Aggrgate Statistics on Matching Rows

```python
import pandas as pd
names = ['Bob','Jessica','Mary','John','Mel']
grades = [76,95,77,78,99]
bsdegrees = [1,1,0,0,1]
msdegrees = [2,1,0,0,0]
phdegrees = [0,1,0,0,0]

GradeList = zip(names,grades,bsdegrees,msdegrees,phdegrees)
columns = ['Names', 'Grades', 'BS', 'MS', 'PhD']
df = pd.DataFrame(data=GradeList,columns=columns)
df
```

```python
df.loc[df['PhD']==0].count()
```

```python
df.loc[df['PhD']==0]['Grades'].mean()
```

# Sorting Data

```python
# Load date
location="Data/gradedata.csv"
df=pd.read_csv(location)
df.head()
```

```python
df=df.sort_values(by='hours', ascending=0)
df.head()
```

```python
df=df.sort_values(by=['age', 'exercise'],
                 ascending=[True, True])
df.head()
```

# Regression

```python
import statsmodels.formula.api as sm
```

```python
result = sm.ols(formula='grade ~ age + exercise + hours', data = df).fit()
```

```python
result.summary()
```

```python
result = sm.ols(
formula='grade ~ exercise + hours',
data = df).fit()
```

```python
result.summary()
```

```python
# grade = 58.5316 + 1.9162 * hours of study + 0.9892 * hours of exercise
# y = a + b1 * hours of study + b2 * hours of exercise
```

```python
# Regression without Intercept = 
```

```python
result = sm.ols(
formula='grade ~ age + exercise + hours -1',
data = df).fit()
```

```python
result.summary()
```
