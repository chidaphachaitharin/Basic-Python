```python
# Load the dataset
import pandas as pd
location="Data/gradedata.csv"
df=pd.read_csv(location)
df.head()
```

```python
df.mode().transpose()
```

```python
data_types=pd.DataFrame(df.dtypes, 
                        columns = ['Data Type'])
data_types
```

```python
# counting Numbers of missing values by column
missing_data_count=pd.DataFrame(df.isnull().sum(), 
                                columns = ['Missing Values'])
missing_data_count
```

```python
# counting Numbers of present observations by column
present_data_count=pd.DataFrame(df.count(), 
                                columns = ['Present Values'])
present_data_count
```

```python
# counting Numbers of unique observations by column
unique_data_count=pd.DataFrame(columns = ['Unique Values'])
for v in list(df.columns.values):
    unique_data_count.loc[v]=[df[v].unique()]
unique_data_count
```

```python
# finding minimum calue for each column
minimum_values=pd.DataFrame(columns=['Min Values'])
for v in list(df.columns.values):
    minimum_values.loc[v]=[df[v].min()]
minimum_values
```

```python
# finding maximum calue for each column
maximum_values=pd.DataFrame(columns=['Max Values'])
for v in list(df.columns.values):
    maximum_values.loc[v]=[df[v].max()]
maximum_values
```

```python
# Making a report with all Computed lists
pd.concat([present_data_count,
          missing_data_count,
          unique_data_count,
          maximum_values,
          minimum_values], axis=1)
```

# Graph Dataset: Line Plot

```python
import pandas as pd
names = ['Bob','Jessica','Mary','John','Mel']
grades = [76,95,77,78,99]
gender = ['Male','Female','Female','Male','Female']
GradeList = zip(names,grades,gender)
columns = ['Names', 'Grades', 'Gender']
df = pd.DataFrame(data=GradeList,columns=columns)
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df.plot()
displayText = "My annotation"
xloc = 1
yloc = df['Grades'].max()
xtext = 8
ytext = 0
plt.annotate(displayText,
           xy=(xloc, yloc),
           xytext=(xtext, ytext),
           xycoords = ('axes fraction', 'data'),
           textcoords = 'offset points')
```

```python
df.plot()
displayText = "My annotation"
xloc = 1
yloc = df['Grades'].max()
xtext = 8
ytext = -150
plt.annotate(displayText,
           xy=(xloc, yloc),
           arrowprops=dict(facecolor='black', shrink=0.05),
           xytext=(xtext, ytext),
           xycoords = ('axes fraction', 'data'),
           textcoords = 'offset points')
```

# Pie chart

```python
import pandas as pd
names = ['Bob','Jessica','Mary','John','Mel']
abscences = [3,0,1,0,8]
detentions = [2,1,0,0,1]
warnings = [2,1,5,1,2]
#gender = ['Male','Female','Female','Male','Female']
GradeList = zip(names,abscences,detentions,warnings)
columns = ['Names', 'Abscences', 'Detentions', 'Warnings']
df = pd.DataFrame(data=GradeList,columns=columns)
df
```

```python
df['TotalDemerits']=df['Abscences']+df['Detentions']+df['Warnings']
df
```

```python
plt.pie(df['TotalDemerits'],
       labels=df['Names'],
       explode=(0,0,0,0,0.1),
       startangle=90,
       autopct='%1.1f%%')
plt.axis('equal')
plt.show
```
