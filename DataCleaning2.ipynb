```python
import pandas as pd 
df=pd.read_csv("Data/gradedata.csv")
```

```python
df.head()
```

```python
import string
exclude = set(string.punctuation)
def remove_punctation(x):
    try:
        x=''.join(ch for ch in x if ch not in exclude)
    except:
        pass
    return x
df.address = df.address.apply(remove_punctation)
```

```python
import pandas as pd 
df=pd.read_csv("Data/gradedata.csv")
```

```python
def remove_whitespace(x):
    try:
        x= ''.join(x.split())
    except:
        pass
    return x
df.address = df.address.apply(remove_whitespace)
```

```python
from datetime import datetime

def standardize_date(thedate):
    """
    Standardizes a given date into MM/DD/YYYY format.
    
    Args:
        thedate (str): The date string to be standardized.
    
    Returns:
        str: The standardized date in MM/DD/YYYY format, or 'MISSING' or 'Incomplete' if the date is invalid.
    """
    # Convert to string and strip any extra spaces
    thedate = str(thedate).strip()
    
    # If the date is missing or invalid (e.g., "missing", "nan", "")
    if not thedate or thedate.lower() == "missing" or thedate.lower() == "nan":
        return "MISSING"
    
    # If the date contains 'x' (indicating an incomplete or incorrect date)
    if 'x' in thedate.lower():
        return "Incomplete"
    
    # Handle cases where year is "00", we replace it with "19" (assuming 1900s)
    if thedate[0:2] == "00":
        thedate = thedate.replace("00", "19", 1)
    
    # Try parsing as MM/DD/YY or MM/DD/YYYY format
    try:
        # First, attempt to parse MM/DD/YY format
        return datetime.strptime(thedate, '%m/%d/%y').strftime('%m/%d/%Y')
    except ValueError:
        pass
    
    # Try parsing as YYYY-MM-DD format
    try:
        return datetime.strptime(thedate, '%Y-%m-%d').strftime('%m/%d/%Y')
    except ValueError:
        pass
    
    # Try parsing DD/MM/YYYY format (common in Europe)
    try:
        return datetime.strptime(thedate, '%d/%m/%Y').strftime('%m/%d/%Y')
    except ValueError:
        pass
    
    # If none of the above formats worked, check if the year is < 1900, return "Incomplete"
    try:
        if int(thedate[:4]) < 1900:
            return "Incomplete"
    except ValueError:
        pass
    
    # If no format worked, return "Incomplete"
    return "Incomplete"
```
