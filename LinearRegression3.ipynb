# A - Linear Regression without Libraries

```python
import numpy as np

# Example student age and birthday month
student_age = 18
student_birthday_month = 7

# Generate independent variable (X) and dependent variable (Y)
X = np.arange(1, student_age + 1)
Y = student_birthday_month * X + student_age + np.random.randint(1, student_age, size=student_age)
```

```python
# Fit Regression Line
X_mean = np.mean(X)
Y_mean = np.mean(Y)

b1 = np.sum((X - X_mean) * (Y - Y_mean)) / np.sum((X - X_mean)**2)
bo = Y_mean - b1 * X_mean

print("Slope (b1):", b1)
print("Intercept (bo):", bo)
```

```python
import matplotlib.pyplot as plt

# Plot Scatter Plot
plt.scatter(X, Y, color='blue', label='Data Points')
plt.title('Scatter Plot of Data Points')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (Y)')
plt.legend()
plt.show()
```

```python
# Plot Fitted Regression Line
Y_pred = b1 * X + bo
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, Y_pred, color='green', label='Fitted Regression Line')
plt.title('Fitted Regression Line')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (Y)')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

# Calculate RMSE
RMSE = np.sqrt(mean_squared_error(Y, Y_pred))
print("Root Mean Squared Error (RMSE):", RMSE)
```

# B - Linear Regression using Linear Regression Library

```python
from sklearn.linear_model import LinearRegression

# Create Linear Regression Model
model = LinearRegression()
```

```python
# Reshape X for scikit-learn
X = X.reshape(-1, 1)

# Fit the model
model.fit(X, Y)

# Make Predictions
Y_pred = model.predict(X)
```

```python
# Plot Linear Regression Model
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, Y_pred, color='green', label='Linear Regression Model')
plt.title('Linear Regression Model')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (Y)')
plt.legend()
plt.show()
```

```python
# Print the regressor coefficient
print("Regressor Coefficient:", model.coef_[0])

# Print the intercept coefficient
print("Intercept Coefficient:", model.intercept_)
```

```python
# Calculate RMSE
RMSE_sklearn = np.sqrt(mean_squared_error(Y, Y_pred))
print("Root Mean Squared Error (RMSE) using scikit-learn:", RMSE_sklearn)
```
